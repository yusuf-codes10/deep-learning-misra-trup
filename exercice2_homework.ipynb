## Training the model (building the neural network)

### Build the model archetecture

```python
# here we use the keras model to trrain
model = keras.models.Sequential()

# adding the hidden layers
model.add(keras.layers.Flatten(input_shape = [28, 28]))

# the hidden layers
model.add(keras.layers.Dense(300, activation = "relu"))
model.add(keras.layers.Dropout(0.3))    # randomly dropout 30%

model.add(keras.layers.Dense(100, activation = "relu"))
model.add(keras.layers.Dropout(0.3))    # randomly dropout 30%


model.add(keras.layers.Dense(10, activation = "softmax")) # cuz we expect 10 labels (y values)
```

```python
# build the model with better syntax
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape = [28, 28]),
    keras.layers.Dense(300, activation = "relu"),
    keras.layers.Dense(100, activation = "relu"),
    keras.layers.Dense(10, activation = "softmax")
])
```

```python
model.summary()
```

```python
model.layers
```

### Compile the model

```python
model.compile(loss= "sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

### fitting the model

```python
histroy = model.fit(x_train, y_train, epochs=30, validation_data=(x_valid, y_valid), batch_size=32)
```

```python
# the 10/adam model
import pandas as pd
pd.DataFrame(histroy.history).plot(figsize=(15, 8))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
# graphing the accuraccy and the loss
# Without 30% dropp
import pandas as pd
pd.DataFrame(histroy.history).plot(figsize=(15, 8))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
# graphing the accuraccy and the loss for the new model
# with 30% drop
import pandas as pd
pd.DataFrame(histroy.history).plot(figsize=(15, 8))
plt.grid(True)
plt.title('Loss/accuracy (30% drop)')
plt.gca().set_ylim(0,1)
plt.show()
```

```python
# evaluating the model
model.evaluate(x_test, y_test)
```

```python
# tesinig it with an image
import numpy as np
import matplotlib.pyplot as plt

# Fashion-MNIST class names
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

# grab a single image from the test set
img = x_test[8]  # shape: (28, 28)

# model expects a batch, so reshape to (1, 28, 28)
img_input = img.reshape(1, 28, 28)

# get predictions
predictions = model.predict(img_input)  # shape: (1, 10)
predicted_class = np.argmax(predictions[0])
confidence = predictions[0][predicted_class] * 100

# visualize
plt.imshow(img, cmap="binary")
plt.title(f"Predicted: {class_names[predicted_class]} ({confidence:.1f}%)")
plt.axis("off")
plt.show()

print(f"Predicted: {class_names[predicted_class]}")
print(f"Confidence: {confidence:.1f}%")
```

Testing with external img

```python
from PIL import Image

# load and preprocess your image to match Fashion-MNIST format
img = Image.open("1.jpg").convert("L")  # convert to grayscale
img = img.resize((28, 28))                        # resize to 28x28
img_array = np.array(img) / 255.0                # normalize to [0, 1]

# invert if your image has dark background (MNIST uses white-on-black)
img_array = 1 - img_array

# predict
predictions = model.predict(img_array.reshape(1, 28, 28))
predicted_class = np.argmax(predictions[0])
print(f"Predicted: {class_names[predicted_class]}")
```

consfuiosn matrix

```python
y_prob = model.predict(x_test)
y_classes = y_prob.argmax(axis=-1)
y_classes  # neatr, probability kinda thing
```

```python
confusion_matrix = tf.math.confusion_matrix(y_test, y_classes)
```

```python
import seaborn as sb

# ax = plt.figure(figsize=(8, 6))
fig = sb.heatmap(confusion_matrix, annot=True, fmt='g', cmap='Greens')  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
fig.set_xlabel('Predicted labels')
fig.set_ylabel('True labels')
fig.set_title('Confusion Matrix')
fig.xaxis.set_ticklabels(class_names)
fig.yaxis.set_ticklabels(class_names)
fig.figure.set_size_inches(10, 10)


plt.show()
```
