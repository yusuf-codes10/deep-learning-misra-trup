## Training the model (building the neural network)

### Build the model archetecture

```python
# here we use the keras model to trrain
model = keras.models.Sequential()

# adding the hidden layers
model.add(keras.layers.Flatten(input_shape = [28, 28]))

# the hidden layers
model.add(keras.layers.Dense(300, activation = "relu"))
model.add(keras.layers.Dropout(0.3))    # randomly dropout 30%

model.add(keras.layers.Dense(100, activation = "relu"))
model.add(keras.layers.Dropout(0.3))    # randomly dropout 30%


model.add(keras.layers.Dense(10, activation = "softmax")) # cuz we expect 10 labels (y values)
```

```python
model.summary()
```

```python
model.layers
```

### Compile the model

```python
model.compile(loss= "sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

### fitting the model

```python
histroy = model.fit(x_train, y_train, epochs=30, validation_data=(x_valid, y_valid), batch_size=32)
```

```python
# graphing the accuraccy and the loss
import pandas as pd
pd.DataFrame(histroy.history).plot(figsize=(15, 8))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```
